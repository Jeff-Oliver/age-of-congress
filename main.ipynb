```python
# import dependencies
import pandas as pd
import zipfile
import json
```

```python
# build the dataframe
# open the zip files
with zipfile.ZipFile('data/BioguideProfiles.zip', 'r') as z:
    # list all the JSON files in the zip
    json_files = [f for f in z.namelist() if f.endswith('.json')]
    # read each JSON files and load into a dataframe
    dfs = [pd.read_json(z.open(f), lines=True) for f in json_files]
    # concatenate all dataframes into one
    df = pd.concat(dfs, ignore_index=True)

# save the dataframe to a CSV file
df.to_csv('data/BioguideProfiles.csv', index=False)

# resulting dataframe has many errors, an alternate data source may be needed
```
