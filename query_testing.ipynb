```python
# import dependencies
from pymongo import MongoClient
from datetime import datetime

```

```python
# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['bioguide']
collection = db['profiles']
```

```python
# 1. Find a specific member by bioId
member = collection.find_one({"usCongressBioId": "A000001"})

if member:
    print(f"{member['givenName']} {member['familyName']}")
else:
    print("Member not found")
    # Check if collection has any documents
    count = collection.count_documents({})
    print(f"Total documents in collection: {count}")
    
    # Show a sample document to verify structure
    sample = collection.find_one()
    if sample:
        print("\nSample document keys:", list(sample.keys()))
    else:
        print("Collection is empty")
```
