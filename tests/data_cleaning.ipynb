```python
import pandas as pd
```

The extracted data still has formatting inconsistencies that need to be corrected for analysis

```python
bioguide_df = pd.read_csv('data/processed/bioguide_table.csv')
# Display the first few rows
print(bioguide_df.head(50))
```

```python
# Standardize the data formats
# convert to dash-separated format for tableau
# skip empty, NaN or year-only entries
def standardize_date(date_str):
    if pd.isna(date_str) or len(date_str) <= 4:
        return date_str
    try:
        return pd.to_datetime(date_str).strftime('%Y-%m-%d')
    except:
        return date_str
bioguide_df['birth_date'] = bioguide_df['birth_date'].apply(standardize_date)
bioguide_df['death_date'] = bioguide_df['death_date'].apply(standardize_date)
# Save the cleaned dataset
bioguide_df.to_csv('data/processed/bioguide_table_cleaned.csv', index=False)
```

```python
# Verify the changes
print(bioguide_df.head(10))
```

```python
# Clean the congress_sessions.csv date formats
# Load the dataset to a dataframe
sessions_df = pd.read_csv('data/processed/congress_sessions.csv')

# rename columns
sessions_df.rename(columns={
    'Congress': 'congress_number',
    'Start_Date': 'start_date',
    'End_Date': 'end_date'
}, inplace=True)

# verify column names
print(sessions_df.head(5))
```

```python
# Standardize the date formats in sessions_df
sessions_df['start_date'] = sessions_df['start_date'].apply(standardize_date)
sessions_df['end_date'] = sessions_df['end_date'].apply(standardize_date)
# Save the cleaned dataset
sessions_df.to_csv('data/processed/congress_sessions_cleaned.csv', index=False)
# Verify the changes
print(sessions_df.head(10))
```
