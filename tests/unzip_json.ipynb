```python
# import dependencies
import pandas as pd
import zipfile
import json
from pathlib import Path
```

```python
# Unzip the dataset using zipfile

# Define the path to the zip file and the extraction directory
zip_file_path = 'data\\BioguideProfiles.zip'
# Define the extraction directory
extraction_dir = "D:\\bioguides" # 

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)
```

Initial EDA

```python
# calculate size of the extracted folder
extracted_size = sum(f.stat().st_size for f in Path(extraction_dir).rglob('*') if f.is_file())
print(f'Total size of extracted folder: {extracted_size / (1024 * 1024):.2f} MB')
```

```python
# Make a list of all the json files in the extracted directory
json_files = list(Path(extraction_dir).rglob('*.json'))
print(f'Number of JSON files found: {len(json_files)}')
```

```python
# read the first json file as a sample
with open(json_files[0], 'r') as f:
    sample_data = json.load(f)
print(json.dumps(sample_data, indent=4))
```

```python
# read the a json file as a sample
with open(json_files[-50], 'r') as f:
    sample_data = json.load(f)
print(json.dumps(sample_data, indent=4))
```
